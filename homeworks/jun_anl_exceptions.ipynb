### Задание 1.

Есть файл с протоколом регистраций пользователей на сайте (registrations.txt).<br>
Каждая строка содержит информацию о имени, электронной почте и возрасте человека. <br><br>


Надо проверить данные из файла, для каждой строки:
 - присутсвуют все три поля
 - поле имени содержит только буквы
 - поле email содержит @ и .
 - поле возраст является числом от 10 до 99<br>

В результате проверки нужно сформировать два файла
 - registrations_good.log для правильных данных, записывать строки как есть
- registrations_bad.log для ошибочных, записывать строку и вид ошибки.<br>

Для валидации строки данных написать метод, который может выкидывать исключения:
 - НЕ присутсвуют все три поля: ValueError
 - поле имени содержит НЕ только буквы: NotNameError (кастомное исключение)
 - поле email НЕ содержит @ и .(точку): NotEmailError (кастомное исключение)
 - поле возраст НЕ является числом от 10 до 99: ValueError
Вызов метода обернуть в try-except.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python

with open('/content/drive/MyDrive/SkillBox/registrations_.txt', 'r') as ff:
    try:
        good_log_file = open('/content/drive/MyDrive/SkillBox/registrations_good.log', 'w') # create / open good log file
        bad_log_file = open('/content/drive/MyDrive/SkillBox/registrations_bad.log', 'w')# create / open bad log file
        
        class NotNameError(Exception): # create NotNameError exception
            def __str__(self):
                self.name = 'NotNameError'
                return self.name

        class NotEmailError(Exception): # create NotEmailError exception
            def __str__(self):
                self.name = 'NotEmailError'
                return self.name

        for line in ff: # checking lines in file
            try:
                name, email, age = line.split() # checking for count of values - ValueError
                if name.isalpha() is False: # checking name consists of letters - NotNameError
                    raise NotNameError()
                elif email.rfind('@') == -1 or email.find('.') == -1: # checking email includes @ or . - NotEmailError
                    raise NotEmailError()
                age = int(age)
                if age < 10 or age > 99: # checking 10 < age < 99 - ValueError
                    raise ValueError("ValueError_Age")
            except ValueError as exc:
                if 'unpack' in exc.args[0]:
                    bad_log_file.write(f'Недостаточно значений ({exc}) в {line}')
                else:
                    bad_log_file.write(f'Поле возраст НЕ является числом от 10 до 99 ({exc}) в {line}')
            except NotNameError as exc:
                bad_log_file.write(f'Поле имени содержит НЕ только буквы ({exc}) в {line}')
            except NotEmailError as exc:
                bad_log_file.write(f'Поле email НЕ содержит @ или .(точку) ({exc}) в {line}')
            else:
                good_log_file.write(line) # writing  right data to good log
    except Exception as exc:
        print("Неожиданная ошибка:", exc)
    finally:
        good_log_file.close() # close good log file
        bad_log_file.close()# close bad log file
```
