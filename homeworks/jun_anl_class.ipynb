### Практика

1. Напишите класс `Fraction` для работы с дробями. Пусть дробь в нашем классе предстает в виде `числитель/знаменатель`. Дробное число должно создаваться по запросу `Fraction(a, b)`, где `a` – это числитель, а `b` – знаменатель дроби. 
2. Добавьте возможность сложения (сложения через оператор сложения) для дроби. Предполагается, что операция сложения может проводиться как только между дробями, так и между дробью и целым числом. Результат операции должен быть представлен в виде дроби.
3. Добавьте возможность взятия разности (вычитания через оператор вычитания) для дробей. Предполагается, что операция вычитания может проводиться как только для двух дробей, так и для дроби и целого числа. Результат операции должен быть представлен в виде дроби.
4. Добавьте возможность умножения (умножения через оператор умножения) для дробей. Предполагается, что операция умножения может проводиться как только для двух дробей, так и для дроби и целого числа. Результат операции должен быть представлен в виде дроби.
5. Добавьте возможность приведения дроби к целому числу через стандартную функцию `int()`.
6. Добавьте возможность приведения дроби к числу с плавающей точкой через стандартную функцию `float()`.
7. Создайте дочерний класс `OperationsOnFraction` и добавьте туда собственные методы `getint` и `getfloat`, которые будут возвращать целую часть дроби и представление дроби в виде числа с плавающей точкой соответственно. 
 

```python
class Fraction():
    def __init__(self, a = None, b = None):
        self.a = a
        self.b = b
    def __str__(self):
        return f'{int(self.a)} / {int(self.b)}'
    
    def gcd(self): #поиск общего наибольшего знаменателя
        c, d = abs(self.a), abs(self.b)
        while c != 0 and d != 0:
            if c > d:
                c %= d
            else:
                d %= c
        return c + d

    def __add__(self, other = None): # операция сложения
        new_fraction = Fraction()
        if isinstance(other, Fraction): # проверка, является ли other дробью
            new_fraction.a = self.a * other.b + other.a * self.b
            new_fraction.b = self.b * other.b
        elif isinstance(other, int): # проверка, является ли other целым
            new_fraction.a = self.a + other*self.b
            new_fraction.b = self.b
        else:
            return print('Неверный ввод')
        # найдем общий знаменатель для числителя и знаменателя и поделим их на него
        com_denominator = new_fraction.gcd()
        new_fraction.a /= com_denominator
        new_fraction.b /= com_denominator
        print(f'Sum of {self} and {other} is {new_fraction}')
        return new_fraction
        
    
    def __sub__(self, other = None): # операция вычитания
        new_fraction = Fraction()
        if isinstance(other, Fraction): # проверка, является ли other дробью
            new_fraction.a = self.a * other.b - other.a * self.b
            new_fraction.b = self.b * other.b
        elif isinstance(other, int): # проверка, является ли other целым
            new_fraction.a = self.a - other*self.b
            new_fraction.b = self.b
        else:
            return print('Неверный ввод')
        # найдем общий знаменатель для числителя и знаменателя и поделим их на него
        com_denominator = new_fraction.gcd()
        new_fraction.a /= com_denominator
        new_fraction.b /= com_denominator
        print(f'Sub of {self} and {other} is {new_fraction}')
        return new_fraction

    def __mul__(self, other = None): # операция умножения
        new_fraction = Fraction()
        if isinstance(other, Fraction): # проверка, является ли other дробью
            new_fraction.a = self.a * self.b
            new_fraction.b = self.b * other.b
        elif isinstance(other, int): # проверка, является ли other целым
            new_fraction.a = self.a * other
            new_fraction.b = self.b
        else:
            return print('Неверный ввод')
        # найдем общий знаменатель для числителя и знаменателя и поделим их на него
        com_denominator = new_fraction.gcd()
        new_fraction.a /= com_denominator
        new_fraction.b /= com_denominator
        print(f'Mul of {self} and {other} is {new_fraction}')
        return new_fraction
    
    def get_int(self): # приведение к целому числу
        print(f'int of {self} is {int(self.a / self.b)}')
        return int(self.a / self.b)

    def get_float(self): # приведение к числу с плавающей точкой
        print(f'float of {self} is {float(self.a / self.b)}')
        return float(self.a / self.b)

frac_1 = Fraction(4, 6)
frac_2 = Fraction(6, 12)
num = 2
print(frac_1)
print(frac_2)
sum_fracs = frac_1 + frac_2 # сложение дробей
sum_frac_int = frac_1 + num # сложение дроби и целого числа
sub_fracs = frac_1 - frac_2 # вычитание дробей
sub_frac_int = frac_1 - num # вычитание из дроби целого числа
mul_fracs = frac_1 * frac_2 # умножение дробей
mul_frac_int = frac_1 * num # умножение дроби и целого числа
sum_frac_int.get_int() # операция int
sum_frac_int.get_float() # операция float
```

```python
class OperationsOnFractions(Fraction): # дочерний класс для операций над дробями
    def getint(self): # приведение к целому числу
        print(f'int of {self} is {int(self.a / self.b)}')
        return int(self.a / self.b)

    def getfloat(self): # приведение к числу с плавающей точкой
        print(f'float of {self} is {float(self.a / self.b)}')
        return float(self.a / self.b)
```
