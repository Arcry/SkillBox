```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Загрузка данных

```python
#df = pd.read_csv('\home\Arcry\SkillBox\homeworks/vgsales.csv')
df = pd.read_csv('/Users/Arcry/SkillBox/files/vgsales.csv')
df.head()
```

### Задача 1

Постройте линейный график динамики объема продаж всех игр жанра "Sports" в Японии. Задайте графику размер 12 на 6

```python
df_sports = df[df['Genre'] == 'Sports'].groupby('Year').sum().reset_index()
df_sports.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_sports['Year'], df_sports['Global_Sales'], label='Global_Sales')
plt.xlabel('Year', fontsize=18)
plt.ylabel('Sales, C', fontsize=18)
plt.tick_params(labelsize=18)
plt.legend(fontsize = 14)
plt.title("График динамики объема продаж всех игр жанра Sports в Японии", fontsize=16)
plt.grid()
plt.show()
```

### Задача 2

В одних координатных осях постройте линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире

```python
df_activision = df[df['Publisher'] == 'Activision'].groupby('Year').sum().reset_index()
df_activision.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_activision['Year'], df_activision['Global_Sales'], label='Global Sales')
plt.plot(df_activision['Year'], df_activision['NA_Sales'], label='NA Sales')
plt.plot(df_activision['Year'], df_activision['EU_Sales'], label='EU Sales')
plt.plot(df_activision['Year'], df_activision['JP_Sales'], label='JP Sales')
plt.xlabel('Year', fontsize=18)
plt.ylabel('Sales, C', fontsize=18)
plt.title("Графики динамики объема продаж игр студии Activision", fontsize=16)
plt.tick_params(labelsize=18)
plt.grid()
plt.legend(fontsize = 14)
plt.show()
```

### Задача 3

В цикле (т.е. на разных координатных осях) постройте линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире

```python
regions = ('Global_Sales', 'NA_Sales', 'EU_Sales', 'JP_Sales')
f,ax=plt.subplots(4, 1, sharey=True, figsize=(18,13))

for i, reg in enumerate(regions):
    ax[i].plot(df_activision['Year'], df_activision[reg])
    ax[i].grid()
    ax[i].set_xlabel('Year', fontsize=16)
    ax[i].set_ylabel(reg, fontsize=16)
    ax[i].tick_params(labelsize=16)
    ax[i].yaxis.set_ticks(np.arange(0, 80, 20))
    ax[i].set_title(f'График динамики объема продаж игр студии Activision в {reg}')
plt.subplots_adjust(hspace=0.4)
plt.show()
```

### Задача 6

Определите 5 платформ, для которых в Японии было продано больше всего игр за все время. Проиллюстрируйте величину продаж на столбчатой диаграмме, столбец с самым высоким значением окрасьте зеленым, а с самым низким - красным

```python
df_max_platform = df.groupby('Platform').sum()['Global_Sales'].sort_values(ascending=False).reset_index()
df_max_platform.head()
```

```python
clrs = []
for value in df_max_platform['Global_Sales'][:5]:
    if value == df_max_platform['Global_Sales'][:5].max():
        clrs.append('g')
    elif value == df_max_platform['Global_Sales'][:5].min():
        clrs.append('r')
    else:
        clrs.append('b')
plt.figure(figsize=(12, 6))
patches = plt.bar(df_max_platform['Platform'][:5], df_max_platform['Global_Sales'][:5], color = clrs)
plt.xlabel('Platform', fontsize=18)
plt.ylabel('Sales', fontsize=18)
plt.tick_params(labelsize=18)
plt.title('5 платформ, для которых в Японии было продано больше всего игр', fontsize=16)
plt.grid()
plt.show()
```

### Задача 7

Постройте линейный график динамики количества игр, выпускаемых Nintendo, по годам. С помощью numpy или pandas определите промежуток, в который издатель каждый год выпускал больше 35 игр, и подсветите этот промежуток времени с помощью зеленого полупрозрачного прямоугольника

```python
df_nintendo = df[df['Publisher'] == 'Nintendo'].groupby('Year')['Name'].count().reset_index()
df_nintendo.head()
```

```python
years_35_games = df_nintendo[df_nintendo['Name'] > 35]['Year']
years_35_games
```

```python
plt.figure(figsize=(18, 6))
plt.plot(df_nintendo['Year'], df_nintendo['Name'])
plt.axhline(y = 35, color = 'r', label = '35 games')
plt.axvspan(xmin = years_35_games.min(), xmax = years_35_games.max(), color='g', alpha=0.5)
plt.xlabel('Year', fontsize=18)
plt.ylabel('Sales, C', fontsize=18)
plt.tick_params(labelsize=18)
plt.grid()
plt.legend(fontsize = 14)
plt.xticks(np.arange(1982, 2018, 2))
plt.title('График динамики количества игр, выпускаемых Nintendo, по годам', fontsize=16)
plt.show()
```
