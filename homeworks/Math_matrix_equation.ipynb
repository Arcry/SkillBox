### Решите задачу аппроксимации из домашнего задания 9 модуля (Часть №3) с помощью псевдообратной матрицы, а не градиентного спуска, используя для вычислений sympy. Насколько более точным получилось решение?

```python
x, y, z = symbols('x y z')
```

```python
MSE = 1/4 * ((1*x + 2*y - 5)**2 + \
             (5*x + 3*y - 6)**2 + \
             (2*x + 4*y - 10)**2 + \
             (3*x + 7*y - 8)**2)
```

```python
X = Matrix([[1, 2],\
           [5, 3],\
           [2, 4],\
           [3, 7]])
B = Matrix([5, 6, 10, 8])
```

```python
X_psevd_inv = (X.T * X)**(-1) * X.T #найдем псевдобратную матрицу
X_psevd_inv
```

```python
point = X_psevd_inv * B # найдем решение
point
```

```python
MSE_best = MSE.subs({x: point[0], y: point[1]}) # подставим решение в функцию MSE
MSE_best
```

```python
MSE_diff = [diff(MSE, x), diff(MSE, y), diff(MSE, z)]
step = 0.01
cnt = 0
point = np.array([0, 0])
while MSE.subs({x: point[0], y: point[1]}) > MSE_best:
    print('MSE =', MSE.subs({x: point[0], y: point[1]}))
    grad = np.array([MSE_diff[0].subs({x: point[0], y: point[1]}),\
                    MSE_diff[1].subs({x: point[0], y: point[1]})])
    point = point - step * grad
    cnt += 1
print(cnt)
```

```python
#  При градиентном спуски то же значение MSE, что и при использовании псевдообратной
# матрицы, при шаге 0.01 достигнуто за 546 итераций
```
