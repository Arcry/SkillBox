```python
import re
```

```python
def find_name(x):
    regex = re.findall(r'. \w+. (\w+)', x)
    if regex:
        return regex[0]
    else:
        return '0'
```

```python
re.findall(r'. \w+. (\w+)', 'Hewlett, Mrs. (Mary D Kingcome)')
```

```python
find_name('Hewlett, Mrs. (Mary D Kingcome)')
```

```python
data[data["Sex"] == 'female']['Name'].apply(find_name).value_counts()
```

```python
for i in data['Name']:
    print(find_name(i))
```
