Напишите программу, которая будет запрашивать на вход числа (каждое с новой строки) до тех пор, пока не будет введён ноль (0). На выход должно выводиться второе по величине значение.
<br><br>
*Ограничение: нельзя пользоваться готовой функцией для нахождения максимума (например, `max()`), готовыми функциями и методами сортировки (например, `sort()`,`sorted()`).*
<br><br>
*Формат ввода:* <br>
4<br>
6<br>
7<br>
7<br>
0
<br><br>

*Формат вывода:* <br>
6

**Рекомендации к выполнению задания:** попробуйте реализовать идею решения из предыдущего задания, используя две переменные для хранения максимумов.

````python
nums = []
num = int(input())
while num > 0:
    nums.append(num)
    num = int(input())
max =  nums[0]
for num in nums[1:]: #нахождение первого максимумв
    if int(num) > max:
        max = num
for _ in range(nums.count(max)): # удаление из списка всех значений, равных первому максимуму
    nums.pop(nums.index(max))
max =  nums[0]
for num in nums[1:]: # поиск второго максимумв
    if int(num) > max:
        max = num
print()
print(max)
````

````python
1from math import sqrt, ceil
numb = int(input())
print()
if numb == 2:
    print('Простое число')
else:
    for i in range(2, ceil(sqrt(numb)) + 1):
        if numb % i == 0:
            print('Составное число')
            break
    else:
        print('Простое число')
````

Напишите программу, которая на вход получает число n и считает по нему сумму сумму 1! + 2! + 3! + ... + n!
<br><br>
*Ограничение: нельзя пользоваться готовой функцией `factorial()` из модуля `math`, функцией `sum()` и их аналогами.*

*Формат ввода:* <br>
3<br><br>

*Формат вывода:* <br>
9

````python
n = int(input())
sum = 0
fact = 1
for i in range(1, n + 1):
    for j in range(1, i + 1):
        fact *= j
    sum += fact
    fact = 1
print()
print(sum)
````

Возраст должен запрашиваться после пола и только в том случае, если пол претендента женский. Если пол претендента мужской, программа должна сообщать о том, что он не подходит. Возраст претенденток должен быть от 18 до 35 лет. Если кандидат удовлетворяет требованиям, должно появляться соответствующее сообщение. Всего в команде могут быть только шесть человек. Когда необходимое число набирается, запись закрывается и выводится сообщение «Запись в команду закрыта».


````python
player_count = 0
team = []
while player_count < 6:
    gender = input('Введите свой пол М / Ж: ')
    if gender == 'Ж':
        age = int(input('Введите свой возраст:'))
        if 18 < age < 35:
            team.append(age)
            team.append(gender)
            player_count += 1
            print('Вы приняты в команду')
        else:
            print('Извините, вы не подходите')
    else:
        print('Запись только в женскую команду')
else:
    print('Запись в команду закрыта')

print()
print(team)
````

Напишите программу, которая выводит число пар одинаковых элементов в списке. При этом, если два элемента образовали пару, они не могут вступить в пару с другими элементами. Программа должна запрашивать на вход слова, каждое с новой строки, пока пользователь не введёт пустую строку. 
<br><br>
*Формат ввода:* <br>
Алиса<br>
Алиса<br>
Алиса<br>
<br><br>

*Формат вывода:* <br>
1

**Рекомендации к выполнению задания:** попробуйте использовать проверку на включённость.

````python
word = input()
names = []
names_no_repeat = []
count = 0
while word:
    if names_no_repeat.count(word) == 0:
        names_no_repeat.append(word)
    names.append(word)
    word = input()
for word in names_no_repeat:
    count += names.count(word) // 2
print()
print(count)

````
