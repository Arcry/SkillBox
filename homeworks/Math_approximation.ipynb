# Блок 1.

С помощью symPy найдите полиномы, описывающие данные наборы точек. Затем проведите полное исследование каждого полученного полинома: проверьте четность/нечетность, найдите нули, промежутки знакопостоянства. Постройте их графики. С помощью изменения промежутка для х добейтесь того, чтобы все нули были отображены на графике функции. <br><br><br>
а) $(-4; -4268), (-3; -1227), (-1; -17), (1; 17), (3; 1227), (4; 4268)$<br><br>
б)$ (-4; -16729), (-3; -3999), (-1; 5), (1; 1), (3; 4005), (4; 16735)$


```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

x, y, z = symbols('x y z')
a = symbols('a0, a1, a2, a3, a4, a5, a6, a7, a8, a9')
```

```python
a[:6][::-1]
```

## а) $(-4; -4268), (-3; -1227), (-1; -17), (1; 17), (3; 1227), (4; 4268)$<br><br>

```python
x_y = [(-4, -4268), (-3, -1227), (-1, -17), (1, 17), (3, 1227), (4, 4268)]
eq_lp = []
for num in x_y:
    fun = 0
    for i, a_i in enumerate(a[:6]):
        fun = Add(fun, a_i * num[0]**i)
    eq_lp.append(fun - num[1])
eq_lp
```

```python
coefs = list(nonlinsolve(eq_lp, a[:6][::-1]))[0]
f = 0
for i, a_n in enumerate(coefs[::-1]):
    f = Add(f, x**i *a_n)
f    
```

```python
plt.figure(figsize=(10,8))
x_range = np.linspace(-10, 10, 100)
y_range = [f.subs(x, x_) for x_ in x_range]
plt.scatter(list(i[0] for i in x_y), list(i[1] for i in x_y), c='r')
plt.plot(x_range, y_range)
plt.xlim(-5, 5)
plt.ylim(-5000, 5000)
plt.grid()
```

```python
solve_univariate_inequality(f.evalf() > 0, x) # Функция положительна на интервалах
```

```python
solve_univariate_inequality(f.evalf() < 0, x) # Функция отрицательна на интервалах
```

```python
if f.subs(x, -2) == f.subs(x, 2):
    print('Функция четная')
elif f.subs(x, -2) == -f.subs(x, 2):
    print('Функция нечетная')
else:
    print('Функция не является четной или нечетной')
```

```python
plot(f, (x, -1, 1))
```

```python
solve(f.evalf(), x) # корни функции 
```

```python
from sympy import *
from sympy.plotting import plot
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol('x')
```

## 1.
### Получите значение MSE меньшее 5.

```python
def print_points_ands_function1(sympy_function):
    def function(x_): return float(sympy_function.subs(x, x_))
    
    points_X = np.array([-2, -1, 0, 1, 2, 3, 3.5, 4, 4.5, 5])
    points_Y = np.array([2, -4, 1, 8, 21, 40, 47, 65, 75, 92])
    plt.xlim(-6, 10)
    plt.ylim(-1, 100)
    
    plt.scatter(points_X, points_Y, c='r')
    x_range = np.linspace(plt.xlim()[0], plt.xlim()[1], num=100)
    function_Y = [function(x_) for x_ in x_range]
    plt.plot(x_range, function_Y, 'b')
    plt.show()
    
    MSE = sum([(points_Y[i] - function(points_X[i]))**2 for i in range(len(points_Y))]) / len(points_Y)
    print(f'MSE = {MSE}')
```

```python
f1 = 48 * x**2 - 128 * x + 83
f1
```

```python
print_points_ands_function1(f1.subs(x, x/4.02 + 1.5))
```
