```python
dat = pd.read_csv('/content/drive/MyDrive/SkillBox/ap-northeast-1.csv')
```

```python
dat.head()
```

```python
dat = dat.shift(periods=1) # сдвинем датафрейм на 1 строку 
dat.head()
```

```python
dat.iloc[0] = list(dat.columns) # перенос названия колонок в строку
dat.columns = ['date', 'instance type', 'OS', 'Region', 'Cost'] # зададим названия столбцов
dat_windows = dat[dat['OS'] == 'Windows'] # создадим отдельный датафрейм с данными по Windows
dat_windows['date'] = pd.to_datetime(dat_windows['date']) # преобразуем столбец даты в тип datetime
dat_windows.index = dat_windows['date']
dat_windows.drop(['date'], axis=1, inplace=True) # удвлим столбец дат
dat_windows
```

```python
for i, j in dat_windows.groupby('instance type'): # посмотрим все виды имеющихся архитектур для windows
    print(i)
```

```python
def i_search(inst):
    return 1 if 'i2' in inst or 'i3' in inst  else 0
```

```python
dat_windows['i_instance'] = dat_windows['instance type'].apply(i_search) # определим архитектуры вида i, и для этого сделаем дополнительный столбец i_instance ,
# в которм 1 - архитектура вида i, 0 - другая
dat_windows.head(20)
```

```python
i_inst_dat_wind = dat_windows[dat_windows['i_instance'] == 1]  # выделим в отдельный датафрейм строки с архитектурами вида i
i_inst_dat_wind.groupby('instance type').sum()
```

```python
from pandas.tseries.offsets import DateOffset # функция для создания списка дней для оси Х графиков
def get_dates(data, name):
    date = data[data['instance type'] == name].index[0]
    dates = []
    while date >= data[data['instance type'] == name].index[-1] - DateOffset(days=1):
        dates.append(date)
        date -= DateOffset(days=1)
    return dates

```

```python
dates_ = get_dates(i_inst_dat_wind[i_inst_dat_wind['instance type'] == 'i2.2xlarge'], 'i2.2xlarge') # построим график для одного вида архитектуры
plt.figure(figsize=(30, 10))
plt.plot(i_inst_dat_wind[i_inst_dat_wind['instance type'] == 'i2.2xlarge'].index, i_inst_dat_wind[i_inst_dat_wind['instance type'] == 'i2.2xlarge']['Cost'])
plt.xticks(dates_, labels=[pd.to_datetime(str(date)).strftime("%d.%m.%Y") for date in dates_], rotation='vertical')
plt.grid()
plt.title(f'Prices of i2.2xlarge instance Windows')
plt.show()
```

```python
fig, axes = plt.subplots(10, 1, figsize=(20,50), sharey='row')
row = 0
for name, data in i_inst_dat_wind.groupby('instance type'):
    dates = get_dates(data=data, name=name)
    axes[row].plot(data[data['instance type'] == name].index, data[data['instance type'] == name]['Cost'])
    axes[row].grid()
    axes[row].set_title(f'Prices of {name} instance Windows')
    axes[row].set_xlabel('Dates')
    axes[row].set_ylabel('Cost')
    axes[row].set_xticks(dates)
    axes[row].set_xticklabels([pd.to_datetime(str(date)).strftime("%d.%m.%Y")\
                                           for date in dates], rotation=90)
    row += 1
plt.subplots_adjust(hspace=0.6)
plt.show()

```
